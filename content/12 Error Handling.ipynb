# Error Handling

</div>

Every program eventually encounters unexpected situations — wrong user input, missing files, network failures. Without error handling, Python crashes with a traceback. With it, your code can respond gracefully instead.

Think of error handling as making your code *polite* — instead of crashing, it explains what went wrong.

## `try` / `except`

The basic pattern:

````python
# Without error handling -- this crashes
# age = int(input("Enter your age: "))  # "hello" -> ValueError

# With error handling
try:
    age = int(input("Enter your age: "))
    print(f"You are {age} years old.")
except ValueError:
    print("That's not a valid number!")
````

The `try` block runs normally. If a `ValueError` occurs, execution jumps to `except` — no crash.

## Multiple `except` clauses

Different errors need different responses:

````python
def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("Cannot divide by zero.")
        return None
    except TypeError:
        print("Both arguments must be numbers.")
        return None

print(safe_divide(10, 2))   # => 5.0
print(safe_divide(10, 0))   # Cannot divide by zero.
print(safe_divide(10, "a")) # Both arguments must be numbers.
````

## `else` clause

Runs only when **no** exception occurred — the happy path:

````python
try:
    number = int(input("Enter a number: "))
except ValueError:
    print("Not a valid number.")
else:
    # Only runs if int() succeeded
    print(f"Double your number: {number * 2}")
````

## `finally` clause

Always runs, whether or not an exception occurred — useful for cleanup:

````python
try:
    result = 10 / int(input("Divisor: "))
except (ValueError, ZeroDivisionError) as e:
    print(f"Error: {e}")
else:
    print(f"Result: {result}")
finally:
    print("Done.")  # always prints
````

## Common built-in exceptions

You've already encountered these — now you know how to handle them:

| Exception | When it occurs |
|-----------|----------------|
| `ValueError` | Wrong value for a type (e.g. `int("hello")`) |
| `TypeError` | Wrong type for an operation |
| `IndexError` | List index out of range |
| `KeyError` | Dictionary key not found |
| `FileNotFoundError` | File doesn't exist |

````python
# Seeing them in action
examples = [
    (lambda: int("hello"),    "int('hello')"),
    (lambda: [1,2,3][9],      "[1,2,3][9]"),
    (lambda: {"a":1}["b"],    "{'a':1}['b']"),
    (lambda: open("nope.txt"),"open('nope.txt')"),
]

for fn, label in examples:
    try:
        fn()
    except Exception as e:
        print(f"{label} -> {type(e).__name__}: {e}")
````

## `raise`

Signal an error intentionally with a meaningful message:

````python
def set_age(age):
    if age < 0 or age > 150:
        raise ValueError(f"Age must be between 0 and 150, got {age}")
    return age

try:
    set_age(-5)
except ValueError as e:
    print(f"Invalid input: {e}")
````
