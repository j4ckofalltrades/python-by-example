# Sets

</div>

A **set** is an unordered collection of *unique* values. Sets automatically remove duplicates and support fast membership testing.

Rule of thumb: use a set when you only care about *what* items exist, not their order or count.

## Creating sets

````python
# With curly braces
fruits = {"apple", "banana", "orange"}
print(fruits)

# From a list (deduplicates automatically)
tags = set(["python", "code", "python", "tutorial"])
print(tags)  # "python" appears only once

# Empty set -- must use set(), NOT {}
empty = set()
print(type(empty))  # => <class 'set'>
````

> **Gotcha:** `{}` creates an empty *dictionary*, not a set. Always use `set()` for an empty set.

## Deduplication

The canonical pattern to remove duplicates from a list:

````python
items = ["apple", "banana", "apple", "orange", "banana"]
unique = list(set(items))
print(unique)  # each fruit appears once (order may vary)
````

## Membership testing

`in` is O(1) for sets — much faster than searching a list for large collections:

````python
taken_usernames = {"alice", "bob", "charlie"}

wanted = "alice"
if wanted in taken_usernames:
    print(f"{wanted} is already taken")
else:
    print(f"{wanted} is available")
````

## Set operations

Sets support mathematical operations with intuitive operators:

````python
alice_friends = {"bob", "carol", "dave", "eve"}
bob_friends  = {"alice", "carol", "frank", "dave"}

# Union: everyone in either set
print("All people:", alice_friends | bob_friends)

# Intersection: common friends
print("Common friends:", alice_friends & bob_friends)

# Difference: in alice's but not bob's
print("Alice only:", alice_friends - bob_friends)
````

## Adding and removing elements

````python
s = {1, 2, 3}

s.add(4)
print(s)  # => {1, 2, 3, 4}

s.remove(2)      # raises KeyError if missing
print(s)  # => {1, 3, 4}

s.discard(99)    # safe -- no error if missing
print(s)  # => {1, 3, 4}
````
