# Tuples

</div>

A **tuple** is an ordered, immutable sequence. Once created, it cannot be changed.

Tuples are used for fixed-structure data — coordinates, RGB values, records — where the values shouldn't change accidentally.

## Creating tuples

Use parentheses `()` or just commas. Python sees the commas and creates a tuple.

````python
point = (3, 4)       # with parentheses
color = 255, 128, 0  # commas alone also work

print(point)   # => (3, 4)
print(color)   # => (255, 128, 0)
print(type(point))  # => <class 'tuple'>
````

### Single-element tuple gotcha

`("hello")` is just a string in parentheses — you need a trailing comma:

````python
not_a_tuple = ("hello")   # this is just a string
a_tuple = ("hello",)      # trailing comma makes it a tuple

print(type(not_a_tuple))  # => <class 'str'>
print(type(a_tuple))      # => <class 'tuple'>
````

## Indexing and slicing

Same syntax as lists — leverage what you already know.

````python
point = (10, 20, 30)

print(point[0])    # => 10
print(point[-1])   # => 30
print(point[1:])   # => (20, 30)
````

## Immutability

Tuples cannot be modified after creation. Trying to assign to an index raises a `TypeError`:

````python
point = (3, 4)
try:
    point[0] = 99  # This raises TypeError
except TypeError as e:
    print(f"Error: {e}")
````

This is a feature, not a bug. Immutability signals "this data is fixed" and prevents accidental changes.

## Tuple unpacking

Assign tuple elements to individual variables in one line — a core Python idiom:

````python
point = (3, 4)
x, y = point
print(f"x={x}, y={y}")  # => x=3, y=4

# Swap two variables elegantly
a, b = 10, 20
a, b = b, a
print(f"a={a}, b={b}")  # => a=20, b=10
````

## Tuples as dictionary keys

Because tuples are immutable, they are *hashable* and can be used as dictionary keys. Lists cannot.

````python
# Map grid coordinates to labels
grid = {
    (0, 0): "origin",
    (1, 0): "right",
    (0, 1): "up",
}
print(grid[(0, 0)])  # => origin
````

## Tuple vs list: when to use which?

- **Tuple**: fixed-structure record where position has meaning (`(lat, lon)`, `(r, g, b)`, `(name, age)`)
- **List**: variable-length collection of the same kind of thing (`[score1, score2, ...]`)
