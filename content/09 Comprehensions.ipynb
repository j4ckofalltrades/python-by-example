# Comprehensions

</div>

Comprehensions are shorthand for a very common pattern: loop over something, optionally filter, and collect results.

You've already written many `for` loops that build lists. A comprehension is just a more compact way to write the same thing.

## List comprehensions

**Before** (for loop):

````python
# Square every number from 1 to 5 — the long way
squares = []
for n in range(1, 6):
    squares.append(n ** 2)
print(squares)  # => [1, 4, 9, 16, 25]
````

**After** (comprehension):

````python
# Same result, one line
squares = [n ** 2 for n in range(1, 6)]
print(squares)  # => [1, 4, 9, 16, 25]
````

Syntax: `[expression for item in iterable]`

### With a filter

Add an `if` clause to keep only matching items:

````python
# Only even numbers
evens = [n for n in range(10) if n % 2 == 0]
print(evens)  # => [0, 2, 4, 6, 8]

# Only words longer than 3 characters
words = ["hi", "hello", "hey", "howdy", "ok"]
long_words = [w for w in words if len(w) > 3]
print(long_words)  # => ['hello', 'howdy']
````

## Dict comprehensions

Syntax: `{key_expr: value_expr for item in iterable}`

````python
# Map each word to its length
words = ["apple", "banana", "cherry"]
word_lengths = {word: len(word) for word in words}
print(word_lengths)  # => {'apple': 5, 'banana': 6, 'cherry': 6}

# Invert a dictionary (swap keys and values)
original = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in original.items()}
print(inverted)  # => {1: 'a', 2: 'b', 3: 'c'}
````

## Set comprehensions

Same as list comprehensions but with curly braces — automatically deduplicates:

````python
sentence = "the quick brown fox jumps over the lazy fox"
unique_words = {word for word in sentence.split()}
print(len(unique_words), "unique words")
````

## Readability rule

If the comprehension gets long or needs nested loops, a regular `for` loop is clearer.

> One simple condition: fine as a comprehension.  
> Multiple conditions or nested loops: use a for loop.

````python
# Readable comprehension
prices = {"apple": 1.2, "banana": 0.5, "cherry": 2.0}
cheap = {k: v for k, v in prices.items() if v < 1.5}
print(cheap)
````

````python
# Nested comprehension -- flatten a 2D list
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [cell for row in matrix for cell in row]
print(flat)  # => [1, 2, 3, 4, 5, 6, 7, 8, 9]
# (deeper nesting than this: just use a for loop)
````
